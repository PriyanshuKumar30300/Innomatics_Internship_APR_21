## Mean 
In statistics, mean is a measure of central tendency of a probability distribution along median and mode. It is also referred to as an expected value.

**Formula: ** 


Mean = (Sum of all the observations/Total number of observations)

In the case of a discrete probability distribution of a random variable X, the mean is equal to the sum over every possible value weighted by the probability of that value; that is, it is computed by taking the product of each possible value x of X and its probability P(x) and then adding all these products together.

```python
df=data['Annual_HH_Income']
```

# correlation

Correlation is used to test relationships between quantitative variables or categorical variables. In other words, it’s a measure of how things are related. The study of how variables are correlated is called correlation analysis. It is a statistical measurement which is used to measure the degree in which two random variables are dependent on each other.

Correlation Coefficient:


### correlation without library

```python
X=list(data['Mthly_HH_Expense'])
Y=df
mu_x=st.mean(X)
mu_y=st.mean(Y)
std_x=st.pstdev(X)
std_y=st.pstdev(Y)

su=0
for i in range(len(X)):
    su+= (X[i]-mu_x)*(Y[i]-mu_y)

su=su/(len(X)*std_x*std_y)
print("Correlation Coefficient :",su)
```

### correlation with library

```python
print("Correlation Coefficient :",np.corrcoef(X,Y))
```

### normal distribution

It is a probability distribution that is symmetric about the mean , showing that the data near the mean are in large number of occurrence than data far from the mean . It is also called Gaussian distribution.As the number of data increases ,it behaves like a normal distribution. It depends on mean and variance.

```python
normal=np.random.normal(Y)
sns.distplot(normal)
```

Features of Normal Distribution:

1) It is a bell curve shape figure i.e it is symmetric about the mean.
2) It has the same mean, median and mode.
3) It follows the 68%, 95%, 99.7% probability distribution rules. It means that by moving 1 stdev , 2 stdev , 3 stdev from mean to both the sides

```python
normal_10000 = np.random.normal(loc = 20, scale = 5, size=15000)

sns.distplot(normal_10000)
```

```python
mean_normal= normal_10000.mean()

sigma = normal_10000.std()

one_std_right= mean_normal+sigma
one_std_left= mean_normal-sigma
two_std_right= mean_normal+(2*sigma)
two_std_left= mean_normal-(2*sigma)
three_std_right= mean_normal+(3*sigma)
three_std_left= mean_normal-(3*sigma)
```

```python
plt.figure(figsize=(20,10))
sns.set_style("darkgrid")
sns.distplot(normal_10000)

plt.axvline(mean_normal, color='coral', label='Mean')

plt.axvline(one_std_right, color='yellow', label='Mean + 1SD')
plt.axvline(one_std_left, color='yellow', label='Mean - 1SD')
plt.axvline(two_std_right, color='green', label='Mean + 2SD')
plt.axvline(two_std_left, color='green', label='Mean - 2SD')
plt.axvline(three_std_right, color='blue', label='Mean + 3SD')
plt.axvline(three_std_left, color='blue', label='Mean - 3SD')
plt.legend();
```

Positively Skewed Normal distribution : It is a normal distribution in which more data are clustered around the left tail side and the right tail are more flatten . It is also called right-skewed distribution. In this case mode< median < mean .

Negatively Skewed Normal distribution :It is a normal distribution in which more data are clustered around the right tail side and left tail are more flatten . It is also called left-skewed distribution. In this case mean< median < mode .
